Streaming

```python
model = init_chat_model(
    "llama-3.3-70b-versatile",
    model_provider="groq"
)
```

```python
response = model.invoke("Write me a 200-word paragraph on Artificial Intelligence.")
print(response.content)
```

```python
for chunk in model.stream("Write me a 200-word paragraph on Artificial Intelligence."):
    print(chunk.content, end="|", flush=True)
```

```python
for chunk in model.stream("Why do parrots have colourful feathers?"):
    print(chunk.content, end="", flush=True)    
```

Batch

```python
responses = model.batch([
    "Why do parrots have colourful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"
])
for i, response in enumerate(responses, start=1):
    print(f"Response {i}:")
    print(response.content)
    print("-" * 50)
```
